This analysis &mdash; based on Wyscout and Fbref data from the top 5 leagues for the 2017/2018 season &mdash; proposes Nick Pope as a possible replacement for Petr Čech. ## 2. Gap Analysis
![EPL 2017/2018 final league points correlations](epl_standings_corr_pts.png "EPL 2017/2018 final league points correlations")
![EPL 2017/2018 final league points distributions](epl_standings_corr_distr.png "EPL 2017/2018 final league points distributions")

For the gap analysis, two intuitive metrics have been employed, namely _expected goals_, _xG_, and _expected goals on target_, _xGOT_ (also referred to as _post-shot expected goals_, _PSxG_); I particularly employ their non-penalty versions. _Expected goals_ is a _pre-shot_ metric, which captures the probability of a shot becoming a goal, given the location of the shooting player, the distance angle to the goal, whether it's a header or a kick, and if the latter is the case &mdash; whether it's kicked with the preferred food or not. _Expected goals on target_, on the other hand, is a _post-shot_ metric, which considers only the shots on target and further takes into account the goal-frame location where the shot is placed. Thus, _xGOT_ may be used as a proxy for gauging shot-taking skill as well as shot-stopping skill.

Arsenal's position in the final rankings &mdash; in terms of number of points &mdash; was $6$<sup>th</sup>, but the goal opportunities generated &mdash; according to the _xG_ model implemented as part of this analysis &mdash; placed Arsenal at the $3$<sup>rd</sup> position ($64.3$), just after Manchester City ($72.7$) and Liverpool ($70.6$). Correspondingly, the team's _xGOT_ ranked them as the $2$<sup>nd</sup> best ($78.7$) just after Manchester City ($90.2$).

On the other hand, in terms of _xG allowed_ and _actual goals allowed_, Arsenal ranked at $6$<sup>th</sup> and $8$<sup>th</sup> position in ascending order, respectively. The team's _xGOT allowed_ ranking was $8$<sup>th</sup>. These numbers suggest that the Gunners had a significant room for improvement, with respect to their defensive performance. Particularly, the shots allowed to be taken against the team were not &mdash; cumulatively speaking &mdash; taken from as dangerous locations as they were executed with skill and, more importantly, they failed to be stopped sufficiently well, which exposed a certain lack of demonstrated skill, namely, in the goalkeeping performances.

Employing the _xGOT_ metric and various related statistics, I've identified Nick Pope (Burnley) as the primary recommendation for a Petr Čech replacement. Norber Murara Neto (Valencia CF) comes in as a relatively close second choice. In the following sections, 3 and 4, I shall consider _The Four Corner Model_ aspects of their performance, and also I'll look at derivative statistics based on _xGOT_.

## 3.The Four Corner Model: Nick Pope
### 3.1. Technical / Tactical

Nick Pope possesses a unique combination of archetypes, being a _shot stopper_ as well as a _sweeper keeper_. His outstanding shot-stopping performances are complemented by footballing intuition and understanding of the game. He pushes higher up the pitch and takes the role of a defender, and as such is capable of assisting with build-up play.

In terms of compatibility with Arsenal's defense and defensive style of play, the distribution of the team's defensive line height isn't that dissimilar from Burnley's, and I expect that Pope wouldn't experience major struggles adapting to the team's style of play and vice versa.

Pope's shot-stopping abilities will be quantitatively and comparatively discussed in depth in the respective section 5, dedicated to $xGOT$.

### 3.2. Physical
With a height of $1.92m$ (top $25$<sup>th</sup> percentile among GKs), Pope weighs $76 kg$ (bottom $15$<sup>th</sup> percentile among GKs), which ranks him as not among the strongest (in terms of strength) goalkeepers, but I wouldn't flag this as a concern. It's just something to take a note of, in case Arsenal's coaches and physios would like to follow up on this, when determining Pope's potential conditioning programmes.

In terms of injury proneness, to the best of my knowledge, as of 1 June 2018, Nick Pope hadn't missed a single day in his senior playing career due to an injury, which is a positive indication.

Now, there's an interesting philosophical question here. Prior to the end of July 2018, Pope hadn't had a single day missed due to injury, [according to Transfermarkt](https://www.transfermarkt.com/nick-pope/verletzungen/spieler/192080). However, on 28 July, he injured his shoulder which took him off the pitch for 162 days! I've opted to use the beginning of June 2018 as a reference point for this transfer proposal, and I wouldn't want to speculate whether or not a potential transfer to Arsenal would've prevented Pope's injury (explained by chaos theory or otherwise). I also wouldn't want to use future data when making transfer proposals.

### 3.3. Psychological

It takes a certain degree of determination and bravery for a goalkeeper to effectively serve as a sweeper keeper, when necessary, as well. This is another pro in Nick Pope's book.

### 3.4. Social
I've examined Pope's social media's presence and I am glad to confirm that there haven't been any red flags discovered. Pope's X's (formerly Twitter) [account](https://x.com/Popey1992/with_replies) reveals a caring personality, supporting various humanitarian causes, as well as posting appreciative and supportive content, regarding his teammates. These are skills, characterising servant leaders, which is a strong leader archetype to have in one's team. Additionally, his Instagram [account](https://www.instagram.com/npope29) is focused on his playing career and features only on-the-pitch content.

## 4. Alternatives
### 4.1. Norberto Murara Neto (Valencia CF)
Should a potential Pope deal be deemed infeasible &mdash; regardless of the associated reason &mdash; Valencia's Norberto Murara Neto is a relatively close contender (with respect to to "the rest of the pack") for a recommendation to replace the underperforming Čech. 

It should be noted that Neto did eventually join Arsenal &mdash; in summer 2024 &mdash; on loan from AFC Bournemouth to effectively replace Aaron Ramsdale on Arsenal's bench as a reserve keeper for David Raya.

With a height of $1.90m$ (top $39$<sup>th</sup> percentile among GKs), Neto weighs $84 kg$ (top $36$<sup>th</sup> percentile among GKs), Neto [had missed](https://www.transfermarkt.com/neto/verletzungen/spieler/111819) 74 days in total due to unknown injuries, as of 1 June 2018.

Neto has featured a family-oriented [social media](https://www.instagram.com/netomurara) [presence](https://x.com/neto_murara). It's not a negative observation, but unlike Nick Pope, there is no indication regarding participation in humanitarian causes &mdash; which, of course, isn't evidence of the absence of such participation, but I felt it's important to note it nevertheless.

## 5. Expected goals on target (_xGOT_) prevented

Given a higher number of expected goals from shots on target, in comparison to Petr Čech ($42$ for Pope vs $37$ for Čech, and $1.17$ vs $1.05$, in terms of per-90-minutes statistics), Pope let $12$ goals less than Petr Čech, which is a substantial difference, considering that the total number of shots was $146$ against Pope and $127$ against Čech.

In addition to Pope's superior &mdash; i.e., lower &mdash; purely _goals-conceded_ (_xGOT conceded_ minus _actual goals conceded_) record, his respective _per-90-minutes_ record is also remarkable, $0.32$, in comparison to Čech's $-0.17$. The negative value is an indication that Čech is letting in more _p90_ goals than what is expected "on average", given shots on target, characterised with the same cumulative difficulty. Neto, aged $29$, and Ter Stegen both score $0.26$ on this statistic, whilst De Gea is the only GK who scores higher than Pope &mdash; a whopping $0.42$. However, in comparison to Pope's market value of £8.4M, De Gea's was £58.6M, whilst Ter Stegen's was £50.2M and Neto's was £12.5M.

Neto with a lower number of _xGOT conceded p90_, conceded approximately the same amount of _goals p90_ as Pope.

## 6. Other considerations
### 6.1. Adaptability
When we factor in data points, pertaining to adaptability/compatibility &mdash; such as foreign language skills (originally from Brazil, playing in Spain) and similarity between cultures &mdash; Nick Pope retains the first position in terms of transfer suggestion preference. It should be noted that, at the time, Unai Emery had just joined Arsenal, and such transfer wouldn't have been unrealistic. However, in terms of overall adaptability, Pope is the clear winner because there are no associated uncertainties in this regard. Also, in terms of _number of goals prevented p90/price_ ratio, Pope is way ahead of every other GK in the top 5 leagues.

### 6.2. Age
In terms of age curves &mdash; i.e., the expectation in accordance to which different positions progress over time, depending on age &mdash; Nick Pope's age was $26$ in summer 2018 and Neto's was $29$. Neto's relative maturity gives him a bit of an edge in this regard, but not completely, because this means that Pope had even further room to grow and develop his current abilities to a greater extent. Also, as a relatively younger player, it means, that he could potentially remain a Gunner for a longer period of time and provide more cumulative value over time, comparatively speaking. This includes a potential transfer for a non-trivial fee, if he ever wished to leave Arsenal. Whilst a relatively older player's potential transfer fee could even be £0 upon contact expiry, and said expiry would be closer in time for relatively older players.

### 6.3. National team call-ups
Nick Pope was called up to the English national squad in March 2018 for the matches against Netherlands and Italy, but didn't get to record any playing minutes in either game and remained on the bench on both occasions.

Neto was called up to the 2012 Olympic tournament in Great Britain with the Brazilian national team, and participated in two group games. He was also called up to the 2015 Copa America tournament but didn't participate in any matches.

I view this as an only marginal advantage in Neto's favour.

### 6.4. Financials
Not only Nick Pope managed to record outstanding performances in 2017/2018 but he also offers the best "bang-for-the-buck" trade-off among all goalkeepers across the top 5 leagues. A potential Neto deal would be a good alternative, because &mdash; if we exclude Pope &mdash; Neto did manage to outshine all other goalkeepers from the top 5 leagues.
